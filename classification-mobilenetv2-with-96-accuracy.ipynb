<h2 style="font-family: Verdana">3.1 <span style='color:#43B0C1; font-weight: bold;'>|</span> Augmentasi Data Gambar</h2>  

**Augmentasi Gambar** adalah proses untuk menciptakan gambar baru untuk pelatihan dengan sedikit mengubah nilai pikselnya.  
Karena kita memiliki dataset yang kecil, kita dapat menerapkan beberapa teknik augmentasi gambar untuk mencegah overfitting pada model.  
Teknik-teknik yang akan kita gunakan adalah:  
- Mengubah kecerahan gambar dengan faktor antara 0.67 hingga 1.33  
- Mengubah kontras gambar dengan faktor antara 0.67 hingga 1.33  

[Faktor-faktor ini telah didefinisikan di bagian konfigurasi]

````python
def augment_image(image):
    image = Image.fromarray(np.uint8(image))
    brightness_factor = random.uniform(cfg.BRIGHTNESS[0], cfg.BRIGHTNESS[1])
    contrast_factor = random.uniform(cfg.CONTRAST[0], cfg.CONTRAST[1])
    image = ImageEnhance.Brightness(image).enhance(brightness_factor)
    image = ImageEnhance.Contrast(image).enhance(contrast_factor)
    return image   
````

````python
def open_images(paths, augment=True):
    images = []
    for path in paths:
        image = load_img(path, target_size=(cfg.IMAGE_SIZE, cfg.IMAGE_SIZE))
        if augment:
            image = augment_image(image)
        image = np.array(image)
        image = image / image.max()
        images.append(image)
    return np.array(images)

# Randomly select and plot a few images with augmentation
k = random.randint(0, 5000)
image_paths = list(df.path[k:k + 10])
labels = list(df.label[k:k + 10])
images = open_images(image_paths, augment=True)

plt.rcParams.update({'font.size': 10})
fig = plt.figure(figsize=(20, 8))
for i in range(0, 10):
    fig.add_subplot(2, 5, i + 1)
    plt .imshow(images[i])
    plt.axis('off')
    plt.title(labels[i])
plt.show()
````

<h2 style="font-family: Verdana">3.4 <span style='color:#43B0C1; font-weight: bold;'>|</span> Data Generators</h2>  



`data_generator` adalah fungsi generator yang akan kita gunakan untuk pelatihan. Fungsi ini memiliki 4 parameter input sebagai berikut:
- `df`: Ini adalah dataframe yang berisi path gambar dan labelnya.
- `batch_size` (int): Ukuran batch dari sampel pelatihan.
- `augment` (boolean): Jika bernilai True, teknik augmentasi yang telah didefinisikan sebelumnya akan diterapkan pada gambar. Kita akan mengaturnya menjadi `True` saat pelatihan dan `False` saat validasi.
- `epochs` (int): Jumlah epoch untuk pelatihan.

````python
def data_generator(df, batch_size=cfg.BATCH_SIZE, augment=False, epochs=cfg.EPOCHS):
    for e in range(epochs):
        for x in range(0, len(df), batch_size):
            image_paths = df.path[x:x + batch_size]
            images = open_images(image_paths)  # Use segmented images
            labels = df.label[x:x + batch_size]
            labels = [label_encoder[label] for label in labels]
            yield images, np.array(labels)
````

````python
train_data_generator = data_generator(df_train, batch_size=cfg.BATCH_SIZE, augment=False, epochs=cfg.EPOCHS)
train_steps = int(len(df_train) / cfg.BATCH_SIZE)
val_data_generator = data_generator(df_val, batch_size=cfg.BATCH_SIZE, augment=False, epochs=cfg.EPOCHS)
val_steps = int(len(df_val) / cfg.BATCH_SIZE)
````

<a id="4.1"></a>
<h2 style="font-family: Verdana">4.1 <span style='color:#43B0C1; font-weight: bold;'>|</span> Build</h2>  
Model Architecture inspired by <a href="https://keras.io/api/applications/mobilenet/"><b>MobilenetV2</b></a> 

````python
import tensorflow as tf
from tensorflow.keras import layers, models

class Config:
    IMAGE_SIZE = 224
cfg = Config()

# Load the MobileNetV2 model, excluding the top layers
base_model = tf.keras.applications.MobileNetV2(input_shape=(cfg.IMAGE_SIZE, cfg.IMAGE_SIZE, 3),
                                               include_top=False,
                                               weights='imagenet')

# Freeze the base model
base_model.trainable = False

# Create the model using Functional API
inputs = layers.Input(shape=(cfg.IMAGE_SIZE, cfg.IMAGE_SIZE, 3))
x = base_model(inputs, training=False)  # Set training to False so base model does not change
x = layers.GlobalAveragePooling2D(name='global_avg_pool')(x)
x = layers.Dropout(0.3, name='dropout1')(x)
x = layers.Dense(128, activation='relu', name='dense1')(x)
outputs = layers.Dense(4, activation='softmax', name='final')(x)

model = models.Model(inputs=inputs, outputs=outputs, name='MobileNetV2_classifier')

# Display the model summary
try:
    print("\nComplete Model Summary:")
    model.summary()
except Exception as e:
    print("Error during complete model summary:", str(e))
````

<a id="4.2"></a>
<h2 style="font-family: Verdana">4.2 <span style='color:#43B0C1; font-weight: bold;'>|</span> Compile</h2>  



Fungsi kerugian yang kita gunakan adalah:

$$\mathcal{L}_{cross-entropy} = -\cfrac{1}{N}\sum_{i=1}^N [y_i\text{log}(\hat y_i) + (1-y_i)\text{log}(1-\hat y_i)]$$

Di mana:
- $\hat y$ adalah label yang diprediksi, dan $y$ adalah label sebenarnya.
- $y_i$ adalah sampel ke-$i$ dari $y$ dan $\hat y_i$ adalah sampel ke-$i$ dari $\hat y$.
- $N$ adalah jumlah sampel.

Fungsi kerugian `sparse_categorical_crossentropy` dan `categorical_crossentropy` pada dasarnya merupakan fungsi kerugian yang sama, hanya saja format inputnya berbeda. 
- Jika $y_i$ dalam format **one-hot encoded**, kita menggunakan `categorical_crossentropy`.
- Namun jika $y_i$ dalam format **integer-encoded**, kita menggunakan `sparse_categorical_crossentropy`.

Untuk kasus kita, label dari data generator yang kita gunakan sudah dalam format **integer-encoded**, maka kita menggunakan `sparse_categorical_crossentropy`. 

Dengan menggunakan `sparse_categorical_crossentropy`, kita dapat memudahkan proses pelatihan model kita, karena tidak perlu melakukan konversi label menjadi format one-hot.

````python
class cfg:
    IMAGE_SIZE = 224
    # Data Augmentation
    BRIGHTNESS = (0.64, 1.37)  # (MIN, MAX)
    CONTRAST = (0.64, 1.37)  # (MIN, MAX)
    # Train-Val Split
    TRAIN_VAL_SPLIT = 0.8  # 80% of all the samples are used for training, and the rest for validation
    # Model Training
    BATCH_SIZE = 8
    EPOCHS = 20
    LEARNING_RATE = 0.00008
    
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=cfg.LEARNING_RATE),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
````

<h2 style="font-family: Verdana">4.3 <span style='color:#43B0C1; font-weight: bold;'>|</span> Train</h2>  

Seperti yang telah disebutkan sebelumnya, kita akan menggunakan **'class weights'** untuk mengatasi masalah dataset yang tidak seimbang. Penggunaan fungsi kerugian cross-entropy akan menciptakan bias menuju kelas yang dominan. Oleh karena itu, kita akan menerapkan pengali, atau **'weights'**, untuk menyeimbangkan fungsi kerugian. Ini disebut sebagai **fungsi kerugian berbobot** (weighted loss function):

$$\mathcal{L}_{cross-entropy} = -\cfrac{1}{N}\sum_{i=1}^N [w_p y_i\text{log}(\hat y_i) + w_n (1-y_i)\text{log}(1-\hat y_i)]$$

### Bagaimana Cara Menghitung Bobot $w_p$ dan $w_n$ untuk Setiap Kelas?

Untuk menghitung bobot yang disebutkan, yaitu $w_p$ (weight untuk kelas positif) dan $w_n$ (weight untuk kelas negatif), Anda cukup menghitung proporsi masing-masing kelas dalam dataset. Dengan cara ini, TensorFlow akan menangani sisa perhitungannya.

`class_weights` adalah sebuah dictionary dengan format `{'CLASS':'PROPORTION'}`. Ketika Anda mengatur nilai ini, TensorFlow akan menggunakan bobot ini saat menghitung fungsi kerugian, sehingga meminimalkan bias terhadap kelas yang mendominasi dan memberikan perhatian lebih kepada kelas yang kurang terwakili.

### Cara Menghitung Proporsi Kelas:
1. **Hitung Jumlah Sampel untuk Setiap Kelas**: Misalnya, jika Anda memiliki kelas A dan B, hitung jumlah sampel untuk A dan B.
2. **Hitung Proporsi untuk Setiap Kelas**: Dapatkan proporsi setiap kelas dengan membagi jumlah sampel pada kelas tersebut dengan total jumlah sampel di seluruh dataset.
3. **Tentukan Bobot Kelas**: Sebagai contoh, jika kelas A memiliki proporsi 0.8 dan kelas B memiliki proporsi 0.2, Anda bisa memberikan bobot yang lebih besar pada kelas yang memiliki proporsi lebih kecil untuk menyeimbangkan kerugian.

Dengan menggunakan bobot kelas ini, Anda dapat membantu model belajar dengan lebih baik dari data yang tidak seimbang.
